### Ejercicio 4

Obten el diagrama de caja del precio del billete usando las funciones de matplotlib y reproduciendo la siguiente figura (tendrs que bucear un poco en los parmetros, y usar patch_artist = True, el tamao de los bigotes est ajustado a 3 IQRs):

```python
# Crear el boxplot
plt.figure(figsize=(6, 8))
plt.boxplot(
    df_titanic['fare'].dropna(),  # Eliminamos valores nulos
    vert=True,                   # Caja vertical
    patch_artist=True,           # Permitir colores personalizados en la caja
    showfliers=True,             # Mostrar valores atpicos
    whis=3                       # Bigotes a 3 veces el IQR
)

# Estilizar el boxplot
plt.title("Boxplot Fare", fontsize=16)
plt.xticks([1], ["Fare"])  # Personalizar etiqueta en el eje x
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar la figura
plt.show()
```

### Ejercicio 5

Crea el grfico de frecuencias absolutas y relativas de las ciudades de embarque del Titanic de forma que repodruzcan la siguiente figura:

```python
# Calcular las frecuencias absolutas y relativas
frecuencia_absoluta = df_titanic['embarked'].value_counts()
frecuencia_relativa = df_titanic['embarked'].value_counts(normalize=True) * 100

# Crear la figura y los subgrficos
fig, axes = plt.subplots(2, 1, figsize=(8, 10))

# Grfico de frecuencias absolutas
axes[0].bar(frecuencia_absoluta.index, frecuencia_absoluta.values, color='blue', edgecolor='black')
axes[0].set_title("Frecuencias Absolutas", fontsize=14)
axes[0].set_xlabel("Ciudad Embarque", fontsize=12)
axes[0].set_ylabel("Num. personas", fontsize=12)

# Grfico de frecuencias relativas
axes[1].bar(frecuencia_relativa.index, frecuencia_relativa.values, color=['brown', 'red', 'blue'], edgecolor='black')
axes[1].set_title("Frecuencias Relativas", fontsize=14)
axes[1].set_xlabel("Ciudad Embarque", fontsize=12)
axes[1].set_ylabel("% del total", fontsize=12)

# Ajustar los espacios entre los grficos
plt.suptitle("Ejercicio 5", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Mostrar la figura
plt.show()
```

### Ejercicio 6

Crea el grfico de dispersin entre el precio y la edad, de forma que reproduzca el siguiente grfico (tendrs que emplear el argumento alpa)

### Ejercicio 7

Vamos a obtener el precio medio por edad, de la siguiente forma agrupa por edad y calcula la media de la tarifa, luego haz lo mismo pero calculando la desviacin de estndar. Guarda los resultados en dos variables distintas.

```python
# Calcular el precio medio por edad
precio_medio_por_edad = df_titanic.groupby('age')['fare'].mean()

# Calcular la desviacin estndar de la tarifa por edad
desviacion_tarifa_por_edad = df_titanic.groupby('age')['fare'].std()

# Convertir a DataFrame para una visualizacin ms clara
precio_medio_df = precio_medio_por_edad.reset_index().rename(columns={'fare': 'Precio Medio'})
desviacion_tarifa_df = desviacion_tarifa_por_edad.reset_index().rename(columns={'fare': 'Desviacin Estndar'})

# Mostrar los resultados
print("Precio Medio por Edad:")
print(precio_medio_df)

print("\nDesviacin Estndar de la Tarifa por Edad:")
print(desviacion_tarifa_df)

```

### Ejercicio 8

Considerando las medias como un conjunto de medidas y las desviaciones como el error de las mismas construye un diagrama de errores que reproduzca la siguiente figura, usando () como moneda:

```python
# Crear el grfico de dispersin con barras de error
plt.figure(figsize=(8, 6))
plt.errorbar(
    precio_medio_por_edad.index,  # Eje x: edades
    precio_medio_por_edad.values,  # Eje y: tarifas promedio
    yerr=desviacion_tarifa_por_edad.values,  # Error: desviaciones estndar
    fmt='o',  # Formato de los puntos
    color='blue',  # Color de los puntos
    ecolor='red',  # Color de las barras de error
    elinewidth=1,  # Grosor de las barras de error
    alpha=0.7  # Transparencia
)

# Configurar etiquetas y ttulo
plt.title("Edad vs Tarifa Promedio en el Titanic", fontsize=14)
plt.xlabel("Edad", fontsize=12)
plt.ylabel("Tarifa Promedio ()", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar el grfico
plt.tight_layout()
plt.show()

```
