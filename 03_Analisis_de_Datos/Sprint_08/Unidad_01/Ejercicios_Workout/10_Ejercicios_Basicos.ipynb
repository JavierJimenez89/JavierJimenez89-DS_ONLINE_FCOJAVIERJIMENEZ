### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear subgrficos
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, 1er grfico
plt.plot(x, y1)
plt.title("Seno")

plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, 2do grfico
plt.plot(x, y2)
plt.title("Coseno")

# Ajustar espacio entre subgrficos
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
# Datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# Crear figura y subgrficos
fig, axes = plt.subplots(3, 1, figsize=(6, 8))  # 3 filas, 1 columna

# Grfico del seno
axes[0].plot(x, y_sin, color='red')
axes[0].set_title("Seno")
axes[0].set_ylim(-1, 1)

# Grfico del coseno
axes[1].plot(x, y_cos, color='green')
axes[1].set_title("Coseno")
axes[1].set_ylim(-1, 1)

# Grfico de la tangente
axes[2].plot(x, y_tan, color='blue')
axes[2].set_title("Tangente")
axes[2].set_ylim(-40, 40)  # Limitar para evitar picos infinitos

# Ajustar el espacio entre subgrficos
plt.tight_layout()

# Mostrar la figura
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Lnea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear subgrficos
plt.subplot(1, 2, 1)  # Subgrfico 1: Seno en magenta
plt.plot(x, y1, color='magenta', linestyle='--')  # Lnea discontinua en magenta
plt.title("Seno")
plt.ylim(-1, 1)  # Limitar el rango en y

plt.subplot(1, 2, 2)  # Subgrfico 2: Coseno en cian
plt.plot(x, y2, color='cyan', linestyle='-.')  # Lnea discontinua punteada en cian
plt.title("Coseno")
plt.ylim(-1, 1)  # Limitar el rango en y

# Ajustar los espacios
plt.tight_layout()

# Mostrar la figura
plt.show()
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperblico y coseno hiperblico se obtienen as:
* np.sinh, seno hiperblico;
* np.cosh, coseno hiperblico;


```python

# Datos
x = np.linspace(0, 10, 100)
y_sinh = np.sinh(x)  # Seno hiperblico
y_cosh = np.cosh(x)  # Coseno hiperblico

# Crear la figura y los subgrficos
fig, axes = plt.subplots(2, 1, figsize=(8, 6))  # 2 filas, 1 columna

# Grfico del seno hiperblico
axes[0].plot(x, y_sinh, color='red')
axes[0].set_title("Seno Hiperblico")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sinh(x)")

# Grfico del coseno hiperblico
axes[1].plot(x, y_cosh, color='blue')
axes[1].set_title("Coseno Hiperblico")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cosh(x)")

# Ajustar los mrgenes
plt.tight_layout()

# Mostrar la figura
plt.show()

```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python
# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log1p(x)  # Logaritmo natural de (1 + x)

# Crear figura y subgrficos
fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # 2 filas, 2 columnas

# Subgrfico 1: Seno
axes[0, 0].plot(x, y1)
axes[0, 0].set_title("Seno")

# Subgrfico 2: Coseno
axes[0, 1].plot(x, y2)
axes[0, 1].set_title("Coseno")

# Subgrfico 3: Tangente
axes[1, 0].plot(x, y3)
axes[1, 0].set_title("Tangente")
axes[1, 0].set_ylim(-40, 40)  # Limitar el eje y para evitar picos

# Subgrfico 4: Logaritmo
axes[1, 1].plot(x, y4)
axes[1, 1].set_title("Logaritmo")

# Ajustar los espacios entre subgrficos
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Mostrar la figura
plt.show()
```


### Ejercicio 6: 

Reproduce la figura siguiente empleando slo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros grficos (fila 1). Luego usa el estilo con mtodos (objetos `axes`) para aadir el tercer grfico.

```python
# Datos
x = np.linspace(0, 10, 100)
y_cos = np.cos(x)
y_sqrt = np.sqrt(x)
y_sin = np.sin(x)

# Crear los primeros grficos usando estilo MATLAB
plt.subplot(2, 1, 1)  # Primera fila, un solo grfico
plt.plot(x, y_cos, label="coseno")
plt.plot(x, y_sqrt, label="raz cuadrada", color="orange")
plt.title("Raz Cuadrada y Coseno")
plt.legend()

# Usar estilo basado en objetos para el grfico de la segunda fila
fig, ax = plt.subplots(1, 1, figsize=(8, 4))  # Segunda fila, un solo grfico
ax.plot(x, y_sin, label="seno")
ax.set_title("Seno")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.legend()

# Mostrar ambos grficos
plt.tight_layout()
plt.show()
```
