### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear subgráficos
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, 1er gráfico
plt.plot(x, y1)
plt.title("Seno")

plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, 2do gráfico
plt.plot(x, y2)
plt.title("Coseno")

# Ajustar espacio entre subgráficos
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
# Datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# Crear figura y subgráficos
fig, axes = plt.subplots(3, 1, figsize=(6, 8))  # 3 filas, 1 columna

# Gráfico del seno
axes[0].plot(x, y_sin, color='red')
axes[0].set_title("Seno")
axes[0].set_ylim(-1, 1)

# Gráfico del coseno
axes[1].plot(x, y_cos, color='green')
axes[1].set_title("Coseno")
axes[1].set_ylim(-1, 1)

# Gráfico de la tangente
axes[2].plot(x, y_tan, color='blue')
axes[2].set_title("Tangente")
axes[2].set_ylim(-40, 40)  # Limitar para evitar picos infinitos

# Ajustar el espacio entre subgráficos
plt.tight_layout()

# Mostrar la figura
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Línea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear subgráficos
plt.subplot(1, 2, 1)  # Subgráfico 1: Seno en magenta
plt.plot(x, y1, color='magenta', linestyle='--')  # Línea discontinua en magenta
plt.title("Seno")
plt.ylim(-1, 1)  # Limitar el rango en y

plt.subplot(1, 2, 2)  # Subgráfico 2: Coseno en cian
plt.plot(x, y2, color='cyan', linestyle='-.')  # Línea discontinua punteada en cian
plt.title("Coseno")
plt.ylim(-1, 1)  # Limitar el rango en y

# Ajustar los espacios
plt.tight_layout()

# Mostrar la figura
plt.show()
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperbólico y coseno hiperbólico se obtienen así:
* np.sinh, seno hiperbólico;
* np.cosh, coseno hiperbólico;


```python

# Datos
x = np.linspace(0, 10, 100)
y_sinh = np.sinh(x)  # Seno hiperbólico
y_cosh = np.cosh(x)  # Coseno hiperbólico

# Crear la figura y los subgráficos
fig, axes = plt.subplots(2, 1, figsize=(8, 6))  # 2 filas, 1 columna

# Gráfico del seno hiperbólico
axes[0].plot(x, y_sinh, color='red')
axes[0].set_title("Seno Hiperbólico")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sinh(x)")

# Gráfico del coseno hiperbólico
axes[1].plot(x, y_cosh, color='blue')
axes[1].set_title("Coseno Hiperbólico")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cosh(x)")

# Ajustar los márgenes
plt.tight_layout()

# Mostrar la figura
plt.show()

```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python
# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log1p(x)  # Logaritmo natural de (1 + x)

# Crear figura y subgráficos
fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # 2 filas, 2 columnas

# Subgráfico 1: Seno
axes[0, 0].plot(x, y1)
axes[0, 0].set_title("Seno")

# Subgráfico 2: Coseno
axes[0, 1].plot(x, y2)
axes[0, 1].set_title("Coseno")

# Subgráfico 3: Tangente
axes[1, 0].plot(x, y3)
axes[1, 0].set_title("Tangente")
axes[1, 0].set_ylim(-40, 40)  # Limitar el eje y para evitar picos

# Subgráfico 4: Logaritmo
axes[1, 1].plot(x, y4)
axes[1, 1].set_title("Logaritmo")

# Ajustar los espacios entre subgráficos
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Mostrar la figura
plt.show()
```


### Ejercicio 6: 

Reproduce la figura siguiente empleando sólo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros gráficos (fila 1). Luego usa el estilo con métodos (objetos `axes`) para añadir el tercer gráfico.

```python
# Datos
x = np.linspace(0, 10, 100)
y_cos = np.cos(x)
y_sqrt = np.sqrt(x)
y_sin = np.sin(x)

# Crear los primeros gráficos usando estilo MATLAB
plt.subplot(2, 1, 1)  # Primera fila, un solo gráfico
plt.plot(x, y_cos, label="coseno")
plt.plot(x, y_sqrt, label="raíz cuadrada", color="orange")
plt.title("Raíz Cuadrada y Coseno")
plt.legend()

# Usar estilo basado en objetos para el gráfico de la segunda fila
fig, ax = plt.subplots(1, 1, figsize=(8, 4))  # Segunda fila, un solo gráfico
ax.plot(x, y_sin, label="seno")
ax.set_title("Seno")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.legend()

# Mostrar ambos gráficos
plt.tight_layout()
plt.show()
```
